### Import Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Importing Covid19 dataset

Data is from 22 January 2020 to 30 April 2020

```python
covid_data=pd.read_csv('Covid19_Confirmed_dataset.csv')
covid_data.head(10)
```

#### Let's check the shape of the dataframe

```python
covid_data.shape
```

```python
col=covid_data.columns
col
```

#### Delete the useless columns

Latitude and Longitude are not important features for us here

```python
covid_data.drop(['Lat',"Long"],axis=1,inplace=True)
```

```python
covid_data.head(10)
```

#### Aggregating the rows by the country

```python
covid_data_aggregated=covid_data.groupby("Country/Region").sum()
```

```python
covid_data_aggregated.head()
```

```python
covid_data_aggregated.shape
```

#### Visualizing data related to a country for example China

```python
covid_data_aggregated.loc['China']
```

### Caculating the first derivative of the curve

```python
covid_data_aggregated.loc["China"].diff().plot()
```

#### Find maximum infection rate for all of the countries.

```python
countries=list(covid_data_aggregated.index)
max_infection_rates=[]
for c in countries:
    max_infection_rates.append(covid_data_aggregated.loc[c].diff().max())
max_infection_rates
```

```python
covid_data_aggregated["max_infection_rates"]=max_infection_rates
covid_data_aggregated.head()
```

#### Create a new dataframe with only needed column

```python
covid_data=pd.DataFrame(covid_data_aggregated['max_infection_rates'])
```

```python
covid_data.head()
```

#### Importing the dataset

```python
happiness_report=pd.read_csv("worldwide_happiness_report.csv")
happiness_report.head()
```

#### Drop the useless columns

```python
useless_cols=["Overall rank", "Score", "Generosity", "Perceptions of corruption"]
```

```python
happiness_report.drop(useless_cols,axis=1,inplace=True)
happiness_report.head()
```

#### Changing the indices of the dataframe

```python
happiness_report.set_index("Country or region",inplace=True)
happiness_report.head()
```

#### Join two dataset we have prepared


```python
covid_data.head()
```

```python
covid_data.shape
```

#### World happiness report Dataset

```python
happiness_report.head()
```

```python
happiness_report.shape
```

```python
data=covid_data.join(happiness_report,how="inner")
data.head()
```

#### Correlation matrix

```python
data.corr()
```

#### Visualization of the results

```python
data.head()
```

#### Plotting GDP vs maximum Infection rate

```python
x=data["GDP per capita"]
y=data["max_infection_rates"]
sns.scatterplot(x,y)
```

```python
x=data["GDP per capita"]
y=data["max_infection_rates"]
sns.scatterplot(x,np.log(y))
```

```python
x=data["GDP per capita"]
y=data["max_infection_rates"]
sns.regplot(x,np.log(y))
```
